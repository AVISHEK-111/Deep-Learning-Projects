Libraries that can be used for image processing

1. matplotlib.image
2. Pillow
3. OpenCV(cv2)


```python
#importing the image module from matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
#loading an image through matplotlib.image module

img = mpimg.imread('/content/360_F_290844781_V4hoIL3E291xvY5nEL7NCaWIoCIQxHfI.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
# here 360 x 540 is dimension and 3 is color channel. as it is rgb image

print(img)
```

```python
#displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
```

Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/360_F_290844781_V4hoIL3E291xvY5nEL7NCaWIoCIQxHfI.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
img = mpimg.imread('/content/dog_image_resized.jpg')
img_plot = plt.imshow(img)
plt.show()
```

```python
img.shape
```

Converting RGB images to Grayscale image using OpenCV

```python
#importing OpenCV library
import cv2
```

```python
img = cv2.imread('/content/360_F_290844781_V4hoIL3E291xvY5nEL7NCaWIoCIQxHfI.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
print(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image on other IDE but this is not allowed in Google Colab.

```python
from google.colab.patches import cv2_imshow
```

```python
#displaying a grayscale image
cv2_imshow(grayscale_image)
```

```python
#saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg', grayscale_image)
```
